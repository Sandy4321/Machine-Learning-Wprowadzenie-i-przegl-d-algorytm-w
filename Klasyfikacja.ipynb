```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
dataset = pd.read_csv('class_data.csv')
```

```python
X = dataset[['X_1', 'X_2']].values
y = dataset['y'].values
```

```python
plt.plot(X[:, 0][y==0], X[:, 1][y==0], 'ro')
plt.plot(X[:, 0][y==1], X[:, 1][y==1], 'bo')
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
clf_knn = KNeighborsClassifier(n_neighbors=5)
clf_knn.fit(X_train, y_train)
```

```python
import helpers as hlp
```

```python
hlp.plot_classification(X_train, y_train, clf_knn, title='KNN')
```

```python
for k in [2, 5, 10, 20, 50, 100]:
    clf_knn = KNeighborsClassifier(n_neighbors=k)
    clf_knn.fit(X_train, y_train)
    hlp.plot_classification(X_test, y_test, clf_knn, title='KNN {}'.format(k))
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
clf_lin = LogisticRegression()
clf_lin.fit(X_train, y_train)
hlp.plot_classification(X_train, y_train, clf_lin, title='Logistyczna regresja')
```

```python
clf_lin.coef_ 
```

```python
from sklearn.svm import SVC
```

```python
for g in [0.01, 0.1, 1, 10, 100, 1000]: 
    clf_svm = SVC(kernel='rbf', C=1, gamma=g)
    clf_svm.fit(X_train, y_train)
    hlp.plot_classification(X_train, y_train, clf_svm, title='SVM, gamma={}'.format(g))
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
for d in [1, 2, 4, 8, 20]: 
    clf_tree = DecisionTreeClassifier(max_depth=d)
    clf_tree.fit(X_train, y_train)
    hlp.plot_classification(X_train, 
                            y_train, 
                            clf_tree, 
                            title='Drzewo, max_depth={}'.format(d))
```
